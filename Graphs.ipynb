```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./owid-covid-data.csv")
df

numerical_features = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'life_expectancy']

# Create histograms
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    plt.hist(df[feature], bins=20, edgecolor='k')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./owid-covid-data.csv")
df

numerical_features = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'life_expectancy']

import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df[['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'life_expectancy']].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./owid-covid-data.csv")
df
# Group data by continent and calculate total new cases and new deaths
continent_data = df.groupby('continent')[['new_cases', 'new_deaths']].sum()

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
continent_data.plot(kind='bar', stacked=True)
plt.title('Distribution of New Cases and New Deaths by Continent')
plt.xlabel('Continent')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./owid-covid-data.csv")
df

numerical_features = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'life_expectancy']

# Calculate the total cases by continent
continent_cases = df.groupby('continent')['total_cases'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(continent_cases, labels=continent_cases.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Total Cases by Continent')
plt.show()
```

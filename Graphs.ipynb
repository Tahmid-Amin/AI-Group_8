```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./owid-covid-data.csv")
df

numerical_features = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'life_expectancy']

# Create histograms
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    plt.hist(df[feature], bins=20, edgecolor='k')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("./owid-covid-data.csv")
df
numerical_features = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'life_expectancy']

# Line plot for 'total_cases' over time
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['total_cases'], marker='o', linestyle='-', color='b')
plt.title('Total Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
